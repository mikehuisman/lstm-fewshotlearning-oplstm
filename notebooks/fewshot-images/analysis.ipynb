```python
import matplotlib.pyplot as plt


val_tasks = 600


cosine_grad_files = [x for x in files if "cosine_grad" in x]
euclid_grad_files = [x for x in files if "euclid_grad" in x]
cosine_proto_files = [x for x in files if "cosine_proto" in x]
euclid_proto_files = [x for x in files if "euclid_proto" in x]


for idx, (group, label) in enumerate(zip([cosine_grad_files, cosine_proto_files, euclid_grad_files,  euclid_proto_files],\
                 ["Cosine gradient", "Cosine prototypes", "Euclidean gradient", "Euclidean prototypes"])):
    
    if idx == 0 or idx ==2:
        if idx == 2:
            plt.show()
        fig, axs = plt.subplots(1, 4, figsize=(16,5))
        if idx == 0:
            axs[0].set_ylabel("Cosine similarity")
        elif idx == 2:
            axs[0].set_ylabel("Euclidean distance")
            
        
    print(label)
    val_data = [None for _ in range(5)]
    train_data = [None for _ in range(5)]
    
    for file in group:
        data = np.load(file)

        num_types = data.shape[0]
        num_points = data.shape[1]

        train_iters_per_val = 10000 #mbs * val_after_num_episodes
        nontest_points = num_points - 1200
        num_val_points = nontest_points-320000
        num_train_points = nontest_points - num_val_points

        num_train_intervals = num_train_points // train_iters_per_val


        val_intervals = [(i*train_iters_per_val+i*600, i*train_iters_per_val+i*600+600) for i in range(num_train_intervals+1)]
        train_intervals = [(i*10000+i*600+600, i*10000+i*600+10000+600) for i in range(num_val_points//600-1)]


        for t in range(num_types):
            val = np.array([data[t, val_intervals[i][0]:val_intervals[i][1]] for i in range(len(val_intervals))]) # (num_val_intervals, 600)
            train = np.array([data[t, train_intervals[i][0]:train_intervals[i][1]] for i in range(len(train_intervals))]) # (num_train_intervals, 600)

            if val_data[t] is None:
                val_data[t] = val
            else:
                val_data[t] = np.concatenate([val_data[t], val], axis=1)
                
            if train_data[t] is None:
                train_data[t] = train
            else:
                train_data[t] = np.concatenate([train_data[t], train], axis=1)
      
    
    
    for t in range(num_types):
        val = val_data[t].mean(axis=1)
        train = train_data[t].mean(axis=1)
        
        train_ci = np.array([train_data[t][i].mean() - st.t.interval(0.95, train_data[t].shape[1], loc=np.mean(train_data[t][i]), scale=st.sem(train_data[t][i]))[0] for i in range(train_data[t].shape[0])])
        val_ci = np.array([val_data[t][i].mean() - st.t.interval(0.95, val_data[t].shape[1], loc=np.mean(val_data[t][i]), scale=st.sem(val_data[t][i]))[0] for i in range(val_data[t].shape[0])])
        
        if t == 0 or not "proto" in label:
            axs[t].plot(np.arange(len(val)), val, label=label, color='blue' if idx%2==0 else 'red', linewidth=1)
            axs[t].fill_between(np.arange(len(val)), val-val_ci, val+val_ci, alpha=0.2, color='blue' if idx%2==0 else 'red', linewidth=1)
            axs[t].set_xlabel("Time")
            
            if idx % 2 != 0:
                axs[t].legend()                    
        else:
            axs[3].plot(np.arange(len(val)), val, label=label, color='blue' if idx%2==0 else 'red', linewidth=1)
            axs[3].fill_between(np.arange(len(val)), val-val_ci, val+val_ci, alpha=0.2, color='blue' if idx%2==0 else 'red', linewidth=1)
            axs[3].set_xlabel("Time")
            
            if idx % 2 != 0:
                axs[3].legend()
                axs[3].set_xlabel("Time")
            break
    
    

    
plt.show()  
```
